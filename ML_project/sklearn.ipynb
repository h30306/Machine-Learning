```python
import pandas as pd
```

```python
data = pd.read_csv('german.txt')
```

```python
data.head()
```

```python
data.Creditability.value_counts()
```

```python
# 分割
from sklearn.cross_validation import train_test_split
```

```python
y = data.Creditability
y
```

```python
# ,1 →直行  ,0→橫列
x = data.drop('Creditability',1)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
x_train.head()
```

```python
# 建模
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression().fit(x_train, y_train)
```

```python
# 每個係數
lr_model.coef_
```

```python
# 截距
lr_model.intercept_
```

```python
# 預測
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression().fit(x_train, y_train)
y_pred = lr_model.predict(x_test)
roc_auc_score(y_test, y_pred)
```

```python
# 決策樹
from sklearn.tree import DecisionTreeClassifier
lr_model = DecisionTreeClassifier().fit(x_train, y_train)
y_pred = lr_model.predict(x_test)
roc_auc_score(y_test, y_pred)
```

```python
# randomforest
from sklearn.ensemble import RandomForestClassifier
lr_model = RandomForestClassifier().fit(x_train, y_train)
y_pred = lr_model.predict(x_test)
roc_auc_score(y_test, y_pred)
```

```python
# adaboost
from sklearn.ensemble import AdaBoostClassifier
lr_model = AdaBoostClassifier().fit(x_train, y_train)
y_pred = lr_model.predict(x_test)
roc_auc_score(y_test, y_pred)
```

```python
# K-fold(分割)
from sklearn import cross_validation
num_folds = 10
num_instances = len(x)
seed = 7
kfold = cross_validation.KFold(n=num_instances,n_folds=num_folds,random_state=seed)
```

```python
model = DecisionTreeClassifier()
results = cross_validation.cross_val_score(model, x, y, cv=kfold)
print(results)
print(results.mean())
```

```python
model = RandomForestClassifier()
results = cross_validation.cross_val_score(model, x, y, cv=kfold)
print(results)
print(results.mean())
```

```python
model = AdaBoostClassifier()
results = cross_validation.cross_val_score(model, x, y, cv=kfold)
print(results)
print(results.mean())
```
